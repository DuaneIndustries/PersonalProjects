```python
df.drop_duplicates(subset='name', inplace=True)
df.count()
```

```python
df.value_counts('types')
```

```python
df[['CardType', 'Subtype']] = df['types'].str.split(" - ", expand = True)
df.head(5)
```

```python
fig = px.scatter(df, x="mana_value", y="CardType", color="color_identity", size = "rarity",
                  hover_data=['name'])
fig.show()
```

```python
df['rarity'].value_counts()
```

```python
rarity_to_num = {'basic': 0, 'token' : 1, 'common' : 2,'uncommon' : 3, 'rare' : 4, 'mythic' : 5}

df['rarity_value'] = df['rarity'].map(rarity_to_num)

df.head(5)

```

```python
df['color_identity'].fillna('X', inplace=True)
```

```python
fig = px.scatter(df, x="mana_value", y="CardType", color="color_identity", size='rarity_value',
                  hover_data=['name'])
fig.show()
```
