```python
df = pd.read_csv('Sample Roasting Data.csv')
df.head(5)
```

```python
dft = df.T
```

```python
dft
```

```python
dft.reset_index(inplace=True)
```

```python
dft.rename(columns={'index' : 'Temp', 0 : 'Colombia-1', 1 : 'Colombia-2', 2 : 'Ethiopia-1', 3 : 'Ethiopia-2', 4 : 'Ethiopia-3', 5 : 'Brazil-1', 6: ' Brazil-2'}, inplace=True )
```

```python
dft.drop([0,1,2,14],inplace=True)
```

```python
dft.reset_index
dft
```

```python
dft['RTime'] = np.arange(0,11,1)
dft
```

```python
dft.rename(columns={' Brazil-2' : 'Brazil-2'},inplace=True)
```

```python

thermo = np.arange(380,425,5)
```

```python
fig = go.Figure()

# Create and style traces
fig.add_trace(go.Scatter(x=dft['RTime'], y=dft['Colombia-1'], name='Colombia Batch 1',
                         line=dict(color='indianred', width=4)))
fig.add_trace(go.Scatter(x=dft['RTime'], y=dft['Ethiopia-1'], name = 'Ethiopia Batch 1',
                         line=dict(color='goldenrod', width=4)))
fig.add_trace(go.Scatter(x=dft['RTime'], y=dft['Brazil-1'], name='Brazil Batch 1',
                         line=dict(color='darkslategray', width=4) # dash options include 'dash', 'dot', and 'dashdot'
))
fig.add_trace(go.Scatter(x=dft['RTime'], y=dft['Colombia-2'], name='Colombia Batch 2',
                        line = dict(color='indianred', width=4, dash='dot')))
fig.add_trace(go.Scatter(x=dft['RTime'], y=dft['Ethiopia-2'], name='Ethiopia Batch 2',
                         line = dict(color='goldenrod', width=4, dash='dot')))
fig.add_trace(go.Scatter(x=dft['RTime'], y=dft['Brazil-2'], name='Brazil Batch 2',
                        line=dict(color='darkslategray', width=4, dash='dot')))

# Edit the layout
fig.update_layout(title='Daily Roast Log',
                   xaxis_title='Roast Time (Mins)',
                   yaxis_title='Temperature (degrees F)')


fig.show()
```

```python
wb = pd.read_csv('WholeBeanInventory.csv')
```

```python
fig = go.Figure(go.Bar(x=wb['CoffeeType'], y=wb['Qty Units (12oz bags)'], name='Roasted Inventory',marker = dict(color='dimgrey')))
fig.add_trace(go.Bar(x=wb['CoffeeType'], y=wb['Sales-Daily'], name='Daily Sales', opacity=.5, marker = dict(color='burlywood')))
fig.add_trace(go.Bar(x=wb['CoffeeType'], y=wb['Sales-Weekly'], name='Weekly Sales', opacity=.5, marker = dict(color='darkred')))
fig.add_trace(go.Bar(x=wb['CoffeeType'], y=wb['Sales-Monthly Forecast)'], name='Monthly Sales', opacity=.5, marker = dict(color='moccasin',pattern_shape='/')))

fig.update_layout(title='Roasted Inventory', barmode='overlay')
fig.show()
```
